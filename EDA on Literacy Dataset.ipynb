```python
#Multivariate Analysis-Involves analysis of more than two variables
```

```python
#What is the percentage of the students whose father and mother both are illiterate but yet they dont feel difficulty in reading English?
df5 = df.groupby(['Father.edu','Mother.edu'])['English.is.difficult'].value_counts()
```

```python
df5
#Ans - 795 students
```

```python
#Convert counts to percentages
df5_per=df5.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
```

```python
df5_per
#Ans- 795 students , that is, 39.08% students feel that way
```

```python
#How many boy students are there who feel maths and science both difficult?
df6 = df.groupby(['Maths.is.difficult','Science.is.difficult'])['Gender'].value_counts()
```

```python
df6
#Ans- 582 boys agree that both subjects are difficult
```
